```python
# 코랩을 본인의 Google drive에 접근하게 하기 위한 코
from google.colab import drive
drive.mount('/content/drive')

# =========================================
# 1. 라이브러리 import
# =========================================
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# =========================================
# 2. 데이터 불러오기
# =========================================
# CSV 파일 경로 (코랩에 업로드 후 경로 설정)
df = pd.read_csv('/content/products.csv')

# 데이터 확인
print(df.head())

# =========================================
# 3. 텍스트 데이터 결합
# =========================================
# 상품명 + 카테고리 + 설명을 하나로 합침
df['text'] = df['product_name'] + " " + df['category'] + " " + df['description']

# 결합된 텍스트 확인
print(df['text'].head())

# =========================================
# 4. TF-IDF 벡터화
# =========================================
# 텍스트를 숫자 벡터로 변환
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(df['text'])

# 벡터 크기 확인
print(tfidf_matrix.shape)

# =========================================
# 5. 추천 함수 정의
# =========================================
def recommend_products(query, top_n=5):
    """
    query: 사용자가 입력한 검색어 (예: '아아')
    top_n: 추천할 상품 개수
    """

    # 5-1. 검색어를 TF-IDF 벡터로 변환
    query_vec = vectorizer.transform([query])

    # 5-2. 코사인 유사도 계산
    similarity = cosine_similarity(query_vec, tfidf_matrix)

    # 5-3. 유사도 높은 순으로 정렬
    similar_indices = similarity[0].argsort()[::-1]

    # 5-4. 상위 N개 선택
    top_indices = similar_indices[:top_n]

    # 5-5. 결과 출력
    results = df.iloc[top_indices][['product_name', 'category', 'description']]

    return results

# =========================================
# 6. 테스트 실행
# =========================================
query = "아아"
results = recommend_products(query, top_n=5)

print(f"\n[검색어: {query}] 추천 결과\n")
print(results)
```
